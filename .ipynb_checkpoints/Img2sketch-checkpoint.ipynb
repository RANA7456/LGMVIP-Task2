```python
import cv2
```

```python
img = cv2.imread('OIP.jpg')
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite('gray.jpg',gray)
```

```python
invert = cv2.bitwise_not(gray)
```

```python
cv2.imwrite('invert.jpg',invert)
```

```python
blurr = cv2.GaussianBlur(invert,(21,21),0)
```

```python
cv2.imwrite('blurr.jpg',blurr)
```

```python
inv_blurr = cv2.bitwise_not(blurr)
```

```python
cv2.imwrite('in_bl.jpg',inv_blurr)
```

```python
sketch = cv2.divide(gray,inv_blurr,scale=256.0)
```

```python
cv2.imwrite('sketch.jpg',sketch)
```
